```python
# First import pandas as pd then import the breast cancer data set from sklearn.datasets
import pandas as pd
from sklearn.datasets import load_breast_cancer

# 1 Load the Breast Cancer dataset first
data = load_breast_cancer()

# Create a DataFrame from the data. This was not easy figuring out, but I figured out that I needed to define the data frame first
df = pd.DataFrame(data.data, columns=data.feature_names)

# 2 Report the dataset's structure by using print, naming it, then telling the data frame what to report with df.
print("Dataset Structure")
print("Shape:", df.shape)
print("Size:", df.size)
print("Columns:", df.columns)
print("Data Types:", df.dtypes)

# 3 First 5 rows which can be done with df.head
print("First 5 Rows of Data")
print(df.head())

# 4 Descriptive stats: calculate mean, median, std, min, and max for each feature.
# I use .head() here so it only displays the first 5 features stats.
descriptive_stats = pd.DataFrame({
    "mean": df.mean(),
    "median": df.median(),
    "std": df.std(),
    "min": df.min(),
    "max": df.max()
})

print("Descriptive Statistics")
print(descriptive_stats.head())
```
